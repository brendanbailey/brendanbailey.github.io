# HDBScan Presentation Jupyter Notebook

HDBScan is a unsupervised clustering technique based on DBScan. It's unsupervised because we do not have a target predicted value to fit it to, and it's a clustering technique because it's used to group like data together.

One of the drawbacks of DBScan is it does not do well with clusters of varying density. That's where HDBScan comes in.

Importing relevant libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
import pandas as pd
from sklearn.preprocessing import StandardScaler
import hdbscan
```

```python
%matplotlib inline
```

#### Creating Density Bars 
This is based on Naftali Harris's Density Bar Plot https://www.naftaliharris.com/blog/visualizing-dbscan-clustering/

```python
np.random.RandomState = 807895 
center_blob = np.random.rand(2,300)
center_blob[0] *= 0.75
```

```python
np.random.RandomState = 967218 
left_blob = np.random.rand(2,75)
left_blob[0] -= 1
```

```python
np.random.RandomState = 711182 
right_blob = np.random.rand(2,75)
right_blob[0] += 0.75
```

```python
new_blob = np.concatenate((center_blob, left_blob), axis = 1)
new_blob = np.concatenate((new_blob, right_blob), axis = 1)
```

```python
plt.tick_params( #http://stackoverflow.com/questions/12998430/remove-xticks-in-a-matplot-lib-plot
    axis='both',          
    which='both',      
    bottom='off',      
    top='off',         
    labelbottom='off',
    left='off',
    labelleft = 'off',
    right = 'off')
plt.scatter(new_blob[0],new_blob[1], facecolors='none', edgecolors='black')
plt.savefig("density_bars_initial.png", bbox_inches='tight', pad_inches=0)
```

Now we have our density, I will apply DBScan.

#### DBScan

```python
scaled_blob = StandardScaler().fit_transform(new_blob.T)
```

```python
dbscan = DBSCAN(eps = 0.225, min_samples=4).fit(scaled_blob)
```

```python
def plot_dbscan(X, labels, figure_name):
    unique_labels = np.unique(labels)
    colors = plt.cm.Spectral(np.linspace(0,1, len(unique_labels)))
    for (label, color) in zip(unique_labels, colors):
        class_member_mask = (labels == label)
        n = X[class_member_mask]
        if label == -1:
            plt.plot(n[:,0],n[:,1], 'o', markerfacecolor = 'none')
        else:
            plt.plot(n[:,0],n[:,1], 'o', markerfacecolor = color)
    plt.tick_params( #http://stackoverflow.com/questions/12998430/remove-xticks-in-a-matplot-lib-plot
    axis='both',          
    which='both',      
    bottom='off',      
    top='off',         
    labelbottom='off',
    left='off',
    labelleft = 'off',
    right = 'off')
    plt.savefig(figure_name, bbox_inches='tight', pad_inches=0)
```

```python
plot_dbscan(scaled_blob, dbscan.labels_, "density_bars_db_scan.png")
```

DBScan got the main dense bar that we were aiming for, but there are all these little clusters that were included as well.

#### HDBScan

```python
clusterer = hdbscan.HDBSCAN(min_samples=11, min_cluster_size=10, allow_single_cluster=True)
```

```python
clusterer.fit(scaled_blob)
```

```python
plot_dbscan(scaled_blob, clusterer.labels_, "density_bars_hdbscan.png")
```

Some of the noise is misclustered, but for the most part we have more control on how many clusters are created.


```python
cd_tree = hdbscan.plots.CondensedTree(clusterer._condensed_tree, cluster_selection_method='eom')
cd_tree.plot()
```
